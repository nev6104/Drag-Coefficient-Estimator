```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image
image = cv2.imread('DIVO.jpg')
```

```python
plt.imshow(image)
plt.axis("off")
```

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_image)
plt.axis('off')
```

```python
edgesclear = cv2.Canny(image, 100, 100)
plt.imshow(edgesclear)
plt.axis("off")
```

```python
# Apply Gaussian Blur  to reduce noise
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
```

```python
plt.axis('off')
plt.imshow(blurred_image)
```

```python
# Detect edges using the Canny edge detector
edges = cv2.Canny(blurred_image, 100, 250)
```

```python
plt.axis('off')
plt.imshow(edges)
```

```python
# Read the image
image = cv2.imread('random/volvo c70.png')
```

```python
def bodymasker (image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
    edges = cv2.Canny(blurred_image, 50, 175)
    plt.axis("off")
    return edges
```

```python
bodymasker(image)
plt.imshow(edges)
```

```python
imagepath = "LEFTS/"
saveto = "edgelefts/"
```

```python
for filename in os.listdir(imagepath):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        edges = bodymasker(image = cv2.imread(os.path.join(imagepath, filename)))
        edge_filename = os.path.splitext(filename)[0] + "_edges.jpg"
        cv2.imwrite(os.path.join(saveto, edge_filename), edges)
```

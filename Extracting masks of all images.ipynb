```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagepath = "testing_normal/"
saveto = "testing_outlines/"
```

```python
def bodymasker (image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
    edges = cv2.Canny(blurred_image, 50, 175)
    plt.axis("off")
    return edges
```

```python
def MASSBODYMASKER(imagepath,saveto):
    for filename in os.listdir(imagepath):
        if filename.endswith(".jpg") or filename.endswith(".png"):
            edges = bodymasker(image = cv2.imread(os.path.join(imagepath, filename)))
            edge_filename = os.path.splitext(filename)[0] + "_edges.jpg"
            cv2.imwrite(os.path.join(saveto, edge_filename), edges)
```

```python
MASSBODYMASKER(imagepath,saveto)
```
